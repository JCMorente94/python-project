```python
# define rooms and items

# Characters

player = {
    "state": "alive"
}

monster_1 = {
    "name": "monster1",
    "type": "monster",
    "state": "alive",
}

monster_2 = {
    "name": "monster2",
    "type": "monster",
    "state": "alive",
}

monster_3 = {
    "name": "monster3",
    "type": "monster",
    "state": "alive",
}

# Weapons

knife = {
    "name": "knife",
    "type": "weapon",
    "target": monster_1,
}

pistol = {
    "name": "pistol",
    "type": "weapon",
    "target": monster_2,
}

bow = {
    "name": "bow",
    "type": "weapon",
    "target": monster_3
}

# Furniture

couch = {
    "name": "couch",
    "type": "furniture",
}

piano = {
    "name": "piano",
    "type": "furniture",
}

table = {
    "name": "table",
    "type": "furniture",
}

queen_bed = {
    "name": "queen bed",
    "type": "furniture",
}

closet = {
    "name": "closet",
    "type": "furniture",
}

double_bed = {
    "name": "double bed",
    "type": "furniture",
}

dresser = {
    "name": "dresser",
    "type": "furniture",
}

dining_table = {
    "name": "dining table",
    "type": "furniture",
}

# Doors

door_a = {
    "name": "door a",
    "type": "door",
    "state": "closed",
}

door_b = {
    "name": "door b",
    "type": "door",
    "state": "closed",
}

door_c = {
    "name": "door c",
    "type": "door",
    "state": "closed",
}

door_d = {
    "name": "door d",
    "type": "door",
    "state": "closed",
}

# Keys

key_a = {
    "name": "key for door a",
    "type": "key",
    "target": door_a,
}

key_b = {
    "name": "key for door b",
    "type": "key",
    "target": door_b,
}

key_c = {
    "name": "key for door c",
    "type": "key",
    "target": door_c,
}

key_d = {
    "name": "key for door d",
    "type": "key",
    "target": door_d,
}

# Rooms

game_room = {
    "name": "game room",
    "type": "room",
}

bedroom_1 = {
    "name": "bedroom 1",
    "type": "room",
}

bedroom_2 = {
    "name": "bedroom 2",
    "type": "room",
}

living_room = {
    "name": "living room",
    "type": "room",
}

outside = {
  "name": "outside"
}

all_rooms = [game_room, bedroom_1, bedroom_2, living_room, outside]

all_doors = [door_a, door_b, door_c, door_d]

all_characters = [player, monster_1, monster_2, monster_3]

all_monsters = [monster_1, monster_2, monster_3]

# define which items/rooms are related

object_relations = {
    "game room": [couch, piano, door_a, monster_1],
    "bedroom 1": [queen_bed, table, door_a, door_b, door_c, monster_2],
    "bedroom 2": [double_bed, dresser, closet, door_b, monster_3],
    "living room": [dining_table, door_c, door_d],
    "couch": [knife],
    "piano": [key_a],
    "table": [pistol],
    "closet": [bow],
    "queen bed": [key_b],
    "double bed": [key_c],
    "dresser": [key_d],
    "outside": [door_d],
    "door a": [game_room, bedroom_1],
    "door b": [bedroom_1, bedroom_2],
    "door c": [bedroom_1, living_room],
    "door d": [living_room, outside],
}

# define game state. Do not directly change this dict. 
# Instead, when a new game starts, make a copy of this
# dict and use the copy to store gameplay state. This 
# way you can replay the game multiple times.

INIT_GAME_STATE = {
    "current_room": game_room,
    "items_collected": [],
    "target_room": outside,
}

def close_door():
    for i in all_doors:
        i["state"] = "closed"
        
def revive_characters():
    for i in all_characters:
        i["state"] = "alive"
        
def all_monsters_are_dead():
    all_dead = True
    for i in all_monsters:
        if i["state"] == "alive":
            all_dead = False       
    return all_dead

def linebreak():
    """
    Print a line break
    """
    print("\n\n")

def start_game():
    """
    Start the game
    """
    print("You wake up on a couch and find yourself in a strange house with no windows which you have never been to before. You don't remember why you are here and what had happened before. You feel some unknown danger is approaching and you must get out of the house, NOW!\n")
    close_door()
    revive_characters()
    play_room(game_state["current_room"])

def play_room(room):
    """
    Play a room. First check if the room being played is the target room.
    If it is, the game will end with success. Otherwise, let player either 
    explore (list all items in this room) or examine an item found here.
    """
    game_state["current_room"] = room
    if player["state"] == 'dead':
        print("You died.\n")
        restart = input("Write restart for start again.\n")
        if restart == 'restart':
            start_game()
        else:
            print("The game ended.")
    elif(game_state["current_room"] == game_state["target_room"]):
        if all_monsters_are_dead():
            print("Congrats! You escaped the room!\n")
        else:
            print("You can't escape, you didn't kill all monsters\n")
            print("You come back to living room")
            play_room(living_room)
    else:
        print("You are now in " + room["name"])
        intended_action = input("What would you like to do?\nType 'explore' or 'examine'?\n").strip()
        if intended_action == "exit":
            print("Has salido del juego.\n")
        elif intended_action == "explore":
            explore_room(room)
            play_room(room)
        elif intended_action == "examine":
            examine_item(input("What would you like to examine?\n").strip())
        else:
            print("Not sure what you mean.\n")
            play_room(room)
        linebreak()

def explore_room(room):
    """
    Explore a room. List all items belonging to this room.
    """
    items = [i["name"] for i in object_relations[room["name"]]]
    print("You explore the room. This is " + room["name"] + ".\nYou find " + ", ".join(items))

def get_next_room_of_door(door, current_room):
    """
    From object_relations, find the two rooms connected to the given door.
    Return the room that is not the current_room.
    """
    connected_rooms = object_relations[door["name"]]
    for room in connected_rooms:
        if(not current_room == room):
            return room

def examine_item(item_name):
    """
    Examine an item which can be a door or furniture or a monster.
    First make sure the intended item belongs to the current room.
    Then check if the item is a door. Tell player if key hasn't been 
    collected yet. Otherwise ask player if they want to go to the next
    room. If the item is not a door, then check if it contains keys.
    Collect the key if found and update the game state. At the end,
    play either the current or the next room depending on the game state
    to keep playing.
    """
    current_room = game_state["current_room"]
    next_room = ""
    output = None
    
    for item in object_relations[current_room["name"]]:
        if(item["name"] == item_name):
            output = "You examine " + item_name + ". "
            
            if(item["type"] == "monster"):
                if item["state"]== "alive":
                    for weapon in game_state["items_collected"]:
                        if(weapon["target"]== item):
                            output += "You killed the monster.\n"
                            item["state"] = "dead"
                    if item["state"] == "alive":
                        output += "The monster killed you.\n"
                        player["state"] = "dead"
                else:
                    output += "The monster is already dead. You don't find anything.\n"
            elif(item["type"] == "door" and item["state"] == "open"):
                output += "The door is already open."
                next_room = get_next_room_of_door(item, current_room)
            elif(item["type"] == "door"):
                have_key = False
                for key in game_state["items_collected"]:
                    if(key["target"] == item):
                        have_key = True
                if(have_key):
                    output += "You unlock it with a key you have.\n"
                    item["state"] = "open"
                    next_room = get_next_room_of_door(item, current_room)
                else:
                    output += "It is locked but you don't have the key.\n"
            else:
                if(item["name"] in object_relations and len(object_relations[item["name"]])>0):
                    item_found = object_relations[item["name"]].pop()
                    game_state["items_collected"].append(item_found)
                    output += "You find " + item_found["name"] + "."
                else:
                    output += "There isn't anything interesting about it.\n"
            print(output)
            break

    if(output is None):
        print("The item you requested is not found in the current room.\n")
    
    if(next_room and input("Do you want to go to the next room?\nEnter 'yes' or 'no'\n").strip() == 'yes'):
        play_room(next_room)
    else:
        play_room(current_room)
```
